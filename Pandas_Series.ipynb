```python
s13 = Series([10,11,12,13,14])      # Series with default index
s14 = Series([5,6,7,8,9],index=['a','b','c','d','e'])    # User defined index 
```

```python
s13
```

```python
s14
```

```python
s13.loc[2]    # Since in this case we did not use any user defined index labels both loc and iloc will work for the given index postions
```

```python
s13.iloc[2]
```

```python
s14.iloc[3]   # iloc only accepts default indices if we insert the value to user defined index labels such as 'a','b',.. in this case, it will show TypeError.
```

```python
s14.loc['d']   # index label are used but if we insert index postion in loc it will show TypeError.
```

```python
s13.at[2]   # for s13 series the output remains the same in case of both at and iat 
```

```python
s13.iat[2]      # The default index works and both the functions at and iat runs without any errors.
```

```python
s14.at['d']    # For s14 series, at function will act as loc only accepts index labels as input
```

```python
s14.iat[3]    # and iat acts like iloc that accepts only index postions as input
```
