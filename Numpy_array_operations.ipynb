```python
brr=np.arange(1,10)     # To print an array with range from 1 to 9 because upper limit (10) is exclusive and lower limit (1) is inclusive 
brr
```

```python
brr.reshape(3,3)     
# Returns an array containing the same data with a new shape. (Changes the dimensions of the array with accordance of the given input by the user)
# (but we need to put appropriate parameters as argument, if we try to make a matrix that can't be created with the given elements, error will occur.) -> {as in the next cell}
# Unlike the free function 'numpy.reshape', this method on ndarray allows the elements of the shape parameter to be passed in as separate arguments.
# Also, this code did'nt change the actual shape of brr in the memory(it not saved) as we have not assigned any variable to this function yet.
# So, if we print brr the one dimensional array will be printed from 1 to 9 as that is saved in the memory. (shown in cell nxt to below one)
```

```python
brr.reshape(4,4)    # Here, brr array has only 9 elements so only (3 by 3) or (3 by 1) or (1 by 3) are the only matrices that can be created. 
```

```python
brr  # Here, we printed brr after the reshape function still it gives original 1 dimensional array as brr array was created because reshape is not saved.
```

```python
brr=brr.reshape(3,3)    # Here, we saved the reshaped brr array by assigning it a variable 'brr', so now the shape of brr is actually changed in the memory.
brr                     # Now we printed brr and it gave 2 dimensional array as after reshape function.
```

```python
brr[0]      # print the 1st row of matrix
```

```python
brr[0][:]     # [0] is the index for 1st row of matrix and ':' is for whole row from index 0 to 2 within the row.
# Here, 1st [] is to access a particular row and then 2nd [] is for performing any function within that particular accessed row.
```

```python
brr[0][-3:-1]    # In the 1st row, print from -3 index(means, from 1st element) to -1 which is exclusive so that means upto 2nd element i.e. 2
#Here, [n][a:b] this only works when a & b are in increasing order means b>a which is true above. (won't work in decreasing order) 
```

```python
brr[0][2:1]   # array is empty as decreasing order in parameters is used.
```

```python
brr[0][0:3:1]  
```

```python
brr[0][::-1]     # Here, we reversed the row.
```
