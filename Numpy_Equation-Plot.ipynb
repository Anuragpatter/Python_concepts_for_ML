```python
from pylab import*
x1=0
x2=0
x1g=x1
x2g=x2
error=1
count=0
while error>1.E-5:
  x1=(9-2*x2)/5
  x2=(3*x1-5)/7
  error=abs(x1-x1g)+abs(x2-x2g)
  semilogy(count,error,'r-o')
  x1g=x1
  x2g=x2
  count+=1
print('{0}and{1}'.format(round(x1,2),round(x2,2)))
show()
```
